## 총 세개의 직선 그래프 y=x, y=-x+5, y=2 동시에 그리고 x라벨, y라벨, 제목, legend 함수 이름으로 해보기

```python
x = [0,1,2,3,4,5]
y1 = [0,1,2,3,4,5]
y2 = [5,4,3,2,1,0]
y3 = [2,2,2,2,2,2]
plt.plot(x, y1, label = 'y=x')
plt.plot(x, y2, label = 'y=-x+5')
plt.plot(x, y3, label = 'y=2')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc = 'upper left')
plt.title('Various Functions')
plt.show()
```

## 점선 그래프

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'rv')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y, 'bo')
plt.axis([0, 6, 0, 30])
plt.show()
```

## 점 그래프 x : 1 ~ 5, y : 1 ~ 5 모든 격자에 점찍기

```python
plt.plot(1,1, 'g+')
```

```python
for i in range(1,6):
    for j in range(1,6):
        plt.plot(i,j,'g+')
plt.show()
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.ones(5)
for i in range(5):
    plt.plot(x,y*(i+1),'bs')
plt.show()
```
