# y=x^2, y=x^1.5

```python
x = np.linspace(0,10,30)
y1 = x**2
y2 = x**1.5
plt.plot(x,y1,"bo-", markersize=2, linewidth=8, label='y=x**2')
plt.plot(x,y2,"rs-", label='y=x**1.5')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Various Functions')
plt.legend(loc="upper left")
plt.show()
```

## y = x^2, y=x^1.5 두개의 함수를 하나의 그래프에 그려보기
+ 범위는 0~10
+ 각각 점이 찍히게 (다른 모양)
+ legend에 함수 표시
+ x 라벨은 X, y 라벨은 Y로
+ 그래프의 보이는 범위는 x : -1 ~ 11 , y : -5 ~ 105

```python
x = np.linspace(0, 10, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=12, label="y = x**2")
plt.plot(x, y2, "gs-", linewidth=2, markersize=12, label="y = x**1.5")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

```python
x = np.logspace(-1, 1, 20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x, y1, "bo-", linewidth=2, markersize=5, label="First")
plt.plot(x, y2, "gs-", linewidth=2, markersize=5, label="Second")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.savefig("mplot.pdf")
```

# 여러가지 np.random method들 확인해 보기

## 1. np.random.rand

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(10000)

plt.hist(x, bins=50, label='bin50', density=True)
plt.hist(x, bins=100, label='bin100', density=True )
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('np.random.rand')
plt.legend(loc='upper left')
plt.show()
```

## 2. np.random.randint

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,101,10000)
plt.hist(x, bins=50, label='bin50',density=True)
plt.hist(x, bins=100, label='bin100',density=True)
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('np.random.randint')
plt.legend(loc='lower right')
plt.show()
```

## 3. random.randn

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
plt.hist(x, bins=50, label='bins 50', density=True)
plt.hist(x, bins=100, label='bins 100', density=True)
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend(loc='lower left')
plt.title=('random.randn')
plt.show()
```

## 4. random.standard_normal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.standard_normal(10000)
plt.hist(x, bins=50, label='bins 50', density=True)
plt.hist(x, bins=100, label='bins 100', density=True)
plt.xlabel('X')
plt.ylabel('Probability')
plt.legend()
plt.title=('random.standard_normal')
plt.show()
```

## 5. random.normal

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(10,20,10000)
plt.hist(x, bins=50, label='bins 50', density=True)
plt.hist(x, bins=100, label='bins 100', density=True)
plt.xlabel('X')
plt.ylabel('probability')
plt.legend()
plt.title=('random.normal')
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
shape = 4 * (np.random.rand(50)*20) **2
plt.scatter(x, y, s=shape, c = colors, marker='*', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.rand(50)
shape = np.pi * (np.random.rand(50)*20) **2
plt.scatter(x, y, s=shape, c = colors, marker='*', alpha=0.7)
plt.show()
```
